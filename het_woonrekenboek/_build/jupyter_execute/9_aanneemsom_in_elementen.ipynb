# Aanneemsom in elementen

Hieronder is een voorbeeld gegeven van de elementen van een aanneemsom op elementkostenniveau. Het is niet de bedoeling hier een kostenbestand te geven waar morgen mee kan worden gerekend, maar de grootheden zijn redelijk. De prijzen waar je het hier over hebt zijn directe kosten. Dat betrekent dat er voor je een aaneemsom hebt nog wat opslagen overheen moeten. In de basis is dit de manier waarop je rekent aan bouwkosten. Een begroting op basis van een bestek zal nog veel verder zijn uitgespliotst, maar dat is voor een residuele waardeberekening niet nodig.

Belangrijke constatering is dat de elementkosten niet 1 op 1 samenhangen met het bvo of aantal kubieke meters van een woning. Stel we hebben een rijwoning van 140 m2 gbo en we hebben vooraf vastgesteld dat de directe kosten bijvoorbeeld € 140.000 zijn. Dan zijn de directe kosten per m2 bvo € 1000. Stel die woning heeft twee lagen en een beukmaat (breedte binnen bouwmuren) van 6m1, dan zou het verlengen van de woning met 1 meter, de woning 12m2 groter maken. In een platte benadering zou je kunnen vermoeden dat de directe kosten € 12.000 hoeger worden, Maar dit is absoluut niet het geval. IN de lijst van elementkosten kun je zien waarom dit het geval is. Bij de basiswoning van 140m2 maken we op bijna alle onderdelen kosten. Als de woning groter wordt veranderen niet op alle onderdelen de hoeveelheden. We kunnen verwachten dat op de volgende essentiele punten extra kosten ontstaan (NB naar kosten voor de kopwoning is niet gekeken):
- grondwerk
- bgg_vloer
- heiwerk (mogelijk)
- verdiepingsvloer
- bouwmuur
- afwerking plat_dak
- variabele installatiekosten

Optelling van die extra kosten is veel lager dan het eerdere bedrag van € 1000 per m2.

Als de woning groter wordt gemaakt door de beukmaat groter te maken, veranderen weer andere kostenposten. Een extra laag geeft weer een ander beeld. Een scherpe analyse van kostenoptimalisaties van een gegeven programma is alleen op deze manier te maken. Dan is er nog geen kwaliteitskeuze gemaakt! De kwaliteit van de woning wordt pas beinvloed door bijvoorbeeld:
- hogere elementprijzen
- meer open gevel
