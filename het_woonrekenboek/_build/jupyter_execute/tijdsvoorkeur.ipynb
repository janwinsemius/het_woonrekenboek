
# Tijdsvoorkeur van geld
Het centrale onderwerp is de tijdsvoorkeur of de tijdwaarde van geld. Als de keuze is om vandaag of over 10 jaar € 1000,- te krijgen, zal vrijwel iedereen kiezen voor vandaag. De tijdsvoorkeur van geld is ook in geld uit te drukken: interest. Als je € 1000,- hebt en je kiest er voor om het een tijd niet te gebruiken, door het bij een bank op een spaarrekening te zetten, krijg je rente als vergoeding voor het uitstellen van het gebruik van je geld. Die vergoeding is een een weerspiegeling van de ontwaarding (inflatie) van geld Er is een duidelijke relatie tussen rente en inflatie. Het kan dan ook voorkomen, als is dat zeer zeldzaam, dar rente negarief is.

## Eindwaarde: rente op rente
De eindwaarde is de toekomstige waarde van een bedrag waarover rente wordt genoten inclusief die rente.  
Een belangrijk concept in de financiële rekenkunde is dat daarbij sprake is van rente op rente. Als je geld op de bank zet je laat de rente die je erover krijgt ook op de bank staan, krijg je de volgende periode niet alleen rente over het initiële bedrag, maar ook over de eerder geincasseerde rente. Er is sprake van zogenaamde samengestelde rente.

De basisrekensom is eenvoudig:

&emsp;$5\%$ van $1000 = 50$

&emsp;$1000 + 50 =$ € $1050$

&nbsp;

Het jaar er na is het iets lastiger hoofdrekenen, maar in essentie hetzelfde:

&emsp;$5\%$ van $1050 = 52,50$

&emsp;$1050 + 52,50 =$ € $1102.50$

&nbsp;

En in jaar 3:

&emsp;$5\%$ van $1102.50 = 55.125$

&emsp;$1102.50 + 55.125 =$ € $1157,625$ enz.

&nbsp;

De som van de waarde in een bepaald jaar kan ook in 1 keer worden berekend:

&emsp;start: 1000

&emsp;jaar 1: $1000 * 1,05 = 1050$

&emsp;jaar 2: $ 1000 * 1,05 * 1,05 = 1102,50$

&emsp;jaar 3: $ 1000 * 1,05 * 1,05 * 1,05 = 1157,625$ enz.

&nbsp;

Wiskundig kan het nog wat korter:

&emsp;jaar 0: $1000 * (1,05)^0 = 1000$

&emsp;jaar 1: $1000 * (1,05)^1 = 1050$

&emsp;jaar 2: $1000 * (1,05)^2 = 1102,50$

&emsp;jaar 3: $1000 * (1,05)^3 = 1157,625$

&emsp;jaar 20: $1000 * (1,05)^{20} = 2653.297$

&emsp;jaar n: $1000 * (1,05)^n$

&nbsp;



```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd
years = list(range(1,21))
samengestelde_rente = [1.05**i - 1 for i in years]
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 1, 1)

ax.bar(years, samengestelde_rente)
yticks = mtick.PercentFormatter(1)
ax.yaxis.set_major_formatter(yticks)
ax.yaxis.grid(True)
ax.set_axisbelow(True)

plt.xticks(years)
plt.xlabel("Jaar")
plt.ylabel("Samengestelde rente")
plt.title("Samengestelde rente op basis van 5% per jaar")
plt.show()
```


## Contante Waarde
In de vastgoedpraktijk is vaak de omgekeerde vraag aan de orde. Je wilt weten hoeveel geld je vandaag opzij moet zetten om over 10 jaar een bepaald bedrag te hebben. Of, wat eigenlijk dezelfde vraag is, je wilt weten wat een bedrag dat je over 10 jaar krijgt vandaag waard is. Dat is bijvoorbeeld de essentie van een huurexploitatie. Je krijgt 50 jaar lang huur. Hoeveel kun je vandaag uitgeven aan een huis op basis van die toekomstige huurinkomsten? Door terug te rekenen naar vandaag reken je de contante waarde uit.

De rekensom wordt eigenlijk omgekeerd. Je moet dan ook niet vermenigvuldigen, maar delen.

Als je volgend jaar € 1000,- krijgt, is de waarde daarvan vandaag, met een rente van 5%:  

&emsp;$1000 / 1.05 = 952.38$

&nbsp;

Als je over 2 jaar € 1000,- krijgt, is de waarde daarvan vandaag:

&emsp;$1000 / 1.05 / 1.05 = 907.03$

&nbsp;

Als je over 3 jaar € 1000,- krijgt, is de waarde daarvan vandaag:

&emsp;$1000 / 1.05 / 1.05 / 1.05 = 863.838$

&nbsp;

Ook dat kan weer snel in 1 keer worden uitgerekend:

&emsp;jaar 0: $1000 / (1,05)^0 = 1000$

&emsp;jaar 1: $1000 / (1,05)^1 = 952.38$

&emsp;jaar 2: $1000 / (1,05)^2 = 907.03$

&emsp;jaar 3: $1000 / (1,05)^3 = 863.838$

&emsp;jaar 20: $1000 / (1,05)^{20} = 367.89$

&emsp;jaar n: $1000 / (1,05)^n$

&nbsp;

Met deze simpele basisbeginselen kun je eigenlijk alle rekensommen maken die nodig zijn om grondexploitaties, huurexploitaties of de haalbaarheid van duurzaamheidsinvesteringen te maken. De complexiteit van de grondexploitatie is niet dat de som ingewikkeld is. De complexiteit is dat de som groot is.
