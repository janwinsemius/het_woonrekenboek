
# Tijdsvoorkeur van geld
Het centrale onderwerp is de tijdsvoorkeur of de tijdwaarde van geld. 


Als de keuze is om vandaag of over 10 jaar € 1000,- te krijgen, zal vrijwel iedereen kiezen voor vandaag. Als je het geld pas over 10 jaar krijgt zal je op zijn minst je wensen of behoeften 10 jaar moeten uitstellen. Dat is wat waard. In de tussentijd was het ook mogelijk geweest met een investering rendement te behalen.
Eén ander probleem van wachten op je geld is dat je in de periode dat je wacht rekening moet houden met inflatie. Als alles duurder wordt is je geld straks minder waard. Prijzen zouden natuurlijk kunnen dalen, maar die kans is toch wat kleiner. 

De tijdsvoorkeur van geld is ook in geld uit te drukken: interest. Als je € 1000,- hebt en je kiest er voor om het een tijd niet te gebruiken, door het bij een bank op een spaarrekening te zetten, krijg je rente als vergoeding voor het uitstellen van het gebruik van je geld. Die vergoeding is in economische termen een blenoningt voor de productiefactor kapitaal. Je zou het ook kunnen beschouwen is een vergoeding voor de ontwaarding (inflatie) van geld. Er is een duidelijke relatie tussen rente en inflatie. Het kan dan ook voorkomen, als is dat zeer zeldzaam, dat rente negatief is.

## Eindwaarde: rente op rente
De eindwaarde is de toekomstige waarde van een bedrag waarover rente wordt genoten inclusief die rente.  
Een belangrijk concept in de financiële rekenkunde is dat daarbij sprake is van rente op rente. Als je geld op de bank zet je laat de rente die je erover krijgt ook op de bank staan, krijg je de volgende periode niet alleen rente over het initiële bedrag, maar ook over de eerder geincasseerde rente. Er is sprake van zogenaamde samengestelde rente.

De basisrekensom is eenvoudig:

\begin{align*}
\text{jaar 1:} & \hspace{1cm} 5\% \times 1000 = 50\\
& \hspace{1cm} 1000 + 50 = \text{€} 1050\\
\\
\text{jaar 2:} & \hspace{1cm} 5\% \times 1050 = 52,50\\
& \hspace{1cm} 1050 + 52,50 = \text{€} 1102.50\\
\\
\text{jaar 3:} & \hspace{1cm} 5\% \times 1102.50 =  55.125\\
& \hspace{1cm} 1102.50 + 55.125 = \text{€} 1157,62\\

\end{align*}


De som van de waarde in een bepaald jaar kan ook in 1 keer worden berekend:
\begin{align*}
\text{start:} \hspace{1cm} & 1000 \\
\text{jaar 1:} \hspace{1cm} & 1000 \times 1,05 = 1050 \\
\text{jaar 2:} \hspace{1cm} & 1000 \times 1,05 \times 1,05 = 1102,50 \\
\text{jaar 3:} \hspace{1cm} & 1000 \times 1,05 \times 1,05 \times 1,05 = 1157,625 \text{etc} \\
\end{align*}

Wiskundig kan het nog wat korter:
\begin{align*}
\text{start:} \hspace{1cm} & 1000 \\
\text{jaar 1:} \hspace{1cm} & 1000 \times (1,05)^{1} = 1050 \\
\text{jaar 2:} \hspace{1cm} & 1000 \times (1,05)^{2} = 1102,50 \\
\text{jaar 3:} \hspace{1cm} & 1000 \times (1,05)^{3} = 1157,625 \\
\text{jaar 20:} \hspace{1cm} & 1000 \times (1,05)^{20} = 2653.297 \\
\text{jaar 20:} \hspace{1cm} & 1000 \times (1,05)^{n} = \text{...etc} \\
\end{align*}

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd
years = list(range(0,21))
samengestelde_rente = [1.05**i - 1 for i in years]
eindwaarde_1000 = [1000 * (1+i) for i in samengestelde_rente]
contante_waarde = [1000 / (1+i) for i in samengestelde_rente]

fig, (ax1, ax2, ax3) = plt.subplots(3,1, figsize=(10,10), sharex=True)

yticks = mtick.PercentFormatter(1)

ax1.bar(years, samengestelde_rente)
ax1.set_title('Samengestelde rente op basis van 5% per jaar')
ax1.yaxis.set_major_formatter(yticks)
ax1.yaxis.grid(True)
ax1.set_axisbelow(True)

ax2.bar(years, eindwaarde_1000)
ax2.set_title('Eindwaarde van €1000 op basis van 5% per jaar')
ax2.yaxis.grid(True)
ax2.set_axisbelow(True)

ax3.bar(years, contante_waarde)
ax3.set_title('Contante waarde van €1000 op basis van 5% per jaar')
ax3.yaxis.grid(True)
ax3.set_axisbelow(True)


plt.xticks(years)
plt.xlabel("Jaar")
```


## Contante Waarde
In de vastgoedpraktijk is vaak de omgekeerde vraag aan de orde. Je wilt weten hoeveel geld je vandaag opzij moet zetten om over 10 jaar een bepaald bedrag te hebben. Of, wat eigenlijk dezelfde vraag is, je wilt weten wat een bedrag dat je over 10 jaar krijgt vandaag waard is. Dat is bijvoorbeeld de essentie van een huurexploitatie. Je krijgt 50 jaar lang huur. Hoeveel kun je vandaag uitgeven aan een huis op basis van die toekomstige huurinkomsten? Door terug te rekenen naar vandaag reken je de contante waarde uit.

De rekensom wordt eigenlijk omgekeerd. Je moet dan ook niet vermenigvuldigen, maar delen.

Als je volgend jaar € 1000,- krijgt, is de waarde daarvan vandaag, met een rente van 5%:

\begin{align*}
\text{jaar 1:} & \hspace{1cm} 1000 / 1.05 = 952.38\\
\\
\text{jaar 2:} & \hspace{1cm} 1000 / 1.05 / 1.05 = 907.03\\
\\
\text{jaar 3:} & \hspace{1cm} 1000 / 1.05 / 1.05 / 1.05 = 863.838\\
\end{align*}


Ook dat kan weer snel in 1 keer worden uitgerekend:

\begin{align*}
\text{start:} \hspace{1cm} & 1000 \\
\text{jaar 1:} \hspace{1cm} & 1000 / (1,05)^{1} = 952.38 \\
\text{jaar 2:} \hspace{1cm} & 1000 / (1,05)^{2} = 907.03 \\
\text{jaar 3:} \hspace{1cm} & 1000 / (1,05)^{3} = 863.838 \\
\text{jaar 20:} \hspace{1cm} & 1000 / (1,05)^{20} = 367.89 \\
\text{jaar n:} \hspace{1cm} & 1000 / (1,05)^{n} = \text{...etc} \\
\end{align*}

Met deze simpele basisbeginselen kun je eigenlijk alle rekensommen maken die nodig zijn om grondexploitaties, huurexploitaties of de haalbaarheid van duurzaamheidsinvesteringen te maken. De complexiteit van de grondexploitatie is niet dat de som ingewikkeld is. De complexiteit is dat de som groot is.
